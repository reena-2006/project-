```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix


df = pd.read_csv('road_accident_data.csv')

# Check data columns
print(df.columns)

# Show first few rows of data to understand structure
print(df.head())

# Example: Fill missing values instead of dropping rows (if there are many missing values)
df['Speed_limit'].fillna(df['Speed_limit'].median(), inplace=True)
df['Weather_conditions'].fillna(df['Weather_conditions'].mode()[0], inplace=True)
df['Road_surface_conditions'].fillna(df['Road_surface_conditions'].mode()[0], inplace=True)
df['Urban_or_Rural_Area'].fillna(df['Urban_or_Rural_Area'].mode()[0], inplace=True)

# Select features and target
features = ['Speed_limit', 'Urban_or_Rural_Area', 'Weather_conditions', 'Road_surface_conditions']
df = pd.get_dummies(df, columns=features)  # Convert categorical to numeric

# Target variable
X = df.drop(['Accident_Severity'], axis=1)
y = df['Accident_Severity']  # target

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation: Confusion Matrix and Classification Report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Feature Importance
importances = pd.Series(model.feature_importances_, index=X.columns)
importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Important Features")
plt.show()
```
